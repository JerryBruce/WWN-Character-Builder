```python
import pdfplumber
pdf = pdfplumber.open("../data/raw/WWN_SRD.pdf")
```

```python
page_index = 5
target_bbox = (306, 410, 612, 760)

# Create the visual
img = pdf.pages[page_index].to_image(resolution=100) # Lower res = faster Ctrl+R
img.draw_rect(target_bbox, stroke="red", stroke_width=2)

# Show the image and the text at the same time
print(pdf.pages[page_index].crop(target_bbox).extract_text())
img
```
